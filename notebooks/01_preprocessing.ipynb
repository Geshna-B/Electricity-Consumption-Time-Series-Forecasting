```python
import pandas as pd
import numpy as np
```

```python
file_path = "D:/SEM 6/TSA/Project/data/raw/household_power_consumption.txt"

df = pd.read_csv(
    file_path,
    sep=";",
    na_values="?",
    low_memory=False
)
```

```python
df['DateTime'] = pd.to_datetime(
    df['Date'] + " " + df['Time'],
    format="%d/%m/%Y %H:%M:%S",
    errors='coerce'
)
```

```python
df = df.drop(columns=['Date', 'Time'])
```

```python
df = df.set_index("DateTime")
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df.isna().sum()
df = df.fillna(method='ffill').fillna(method='bfill')
```

```python
df = df.sort_index()
df = df[~df.index.duplicated(keep='first')]
df_hourly = df.resample('H').mean()
```

```python
import os

os.makedirs("data/processed", exist_ok=True)
```

```python
df.to_csv("data/processed/power_clean_minute.csv")
df_hourly.to_csv("data/processed/power_clean_hourly.csv")
```
